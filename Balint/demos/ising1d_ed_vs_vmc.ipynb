# 1D Ising: Exact Diagonalization vs Our VMC Models

This demo compares the exact ground-state energy of the 1D transverse-field Ising model, computed with NetKet exact diagonalization, against short VMC runs using our project-owned `RBM`, `FFNN`, and `CNN` models.

For speed, this notebook uses `L=5` and a very small number of optimization steps. The goal is to verify the pipeline, not to fully converge the models.

## What is project-owned here?

- NetKet provides the temporary Ising operator and exact-diagonalization backend.
- Our code provides the neural-network architectures, variational state wrapper, optimizer API, and VMC driver.
- The helper module in this folder contains the reusable Python code so the notebook stays short and readable.

```python
from ising1d_ed_vs_vmc_helper import run_demo
```

```python
results = run_demo(length=5, transverse_field=1.0)
```

```python
print(f"L = {results['length']}")
print(f"h = {results['transverse_field']}")
print(f"Exact ground-state energy: {results['exact_energy']:.6f}")
print()
print("Final comparison")
for entry in results['results']:
    print(
        f"{entry['model']:>4}: final energy = {entry['final_energy']:.6f}, "
        f"delta to exact = {entry['delta_to_exact']:+.6f}"
    )
```

## Notes

- The printed energy traces show how each model changes over a few optimization steps.
- With the current small sample counts and only a handful of iterations, the models will not be close to fully converged.
- This notebook is mainly a quick integration check that ED, the NetKet adapter, and our own model/VMC stack all work together.
